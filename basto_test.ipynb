```python
import bson
import pandas as pd
```

```python
def bson_data(url):
    with open(url,'rb') as f:
        datos_bson= bson.decode_all(f.read())
    datos= pd.DataFrame.from_records(datos_bson)
    return datos
```

```python
df_animal=bson_data('dump/test/animals.bson')
df_animal.sample(10)
```

```python
row_data= bson_data('dump/test/plots.bson')
row_data.sample(5)
```
